```python
import pandas as pd
from math import ceil
import json

df = pd.read_excel(r"C:\Users\Rafaeros\Documents\Development\Python\data-migration\Estoque.xlsx", sheet_name="Bloco K MARÇO")

df = df[["CODÍGO", "FINALIDADE", "PROPRIETARIO", "SALDO TOTAL", "UNIDADE", "CUSTO UNITARIO", "TIPO/PROPRIETARIO", "DECLARA", "NCM"]]
owners: list[str] = ["F&K GROUP TECNOLOGIA EM SISTEMAS AUTOMOTIVOS LTDA.", "ESTOQUE DEUTSCH", "ITENS OBSOLETOS", "OVERSTOCK"]
df["SALDO TOTAL"] = pd.to_numeric(df["SALDO TOTAL"], errors="coerce")
df = df[df["SALDO TOTAL"] > 0]
df = df[df["PROPRIETARIO"].isin(owners)]

print(df.head())

groups = df.groupby("TIPO/PROPRIETARIO")

for owner_type, group in groups:
    group = group.reset_index(drop=True)
    order_number = ceil(len(group) / 20)
    orders: list[dict] = []

    for i in range(order_number):
        slice = group.iloc[i * 20 : (i + 1) * 20]
        
        orders.append({
            "tipo_proprietario": owner_type,
            "pedido_numero": i + 1,
            "quantidade_itens": len(slice),
            "itens": slice.drop(columns=["TIPO/PROPRIETARIO"]).to_dict(orient="records"),
        })


    json_file: str = owner_type.lower().replace("/", "_") + ".json"
    with open(json_file, "w") as f:
        json.dump(orders, f, indent=4, ensure_ascii=False)

print(f"Total de pedidos: {len(orders)}")
```
